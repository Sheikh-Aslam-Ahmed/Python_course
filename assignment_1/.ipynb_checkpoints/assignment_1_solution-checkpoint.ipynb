3.Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.

```python
first = input("Enter your first Name : ") 
last = input("Enter your last Name : ") 
reverse = first[::-1] + " " + last[::-1]
print(reverse)
```

```python
k = 2
for i in range(1,10): 
    if i < 6:
        for j in range(0,i):
            print("* ", end="")
    else:
       
        for j in range(i-k,0,-1):
            print("* ", end="")
        k = k+2
    print("\r")
```

```python
word = input("Enter your word : ") 
output_word = ""
is_upper = word[0].isupper()
tab_count = 1
for i in word:
    if i.isspace():
        output_word = output_word + i
        continue
    elif not i.isalnum():
        output_word = output_word + i
        continue
    elif i.isupper() != is_upper:
        output_word = output_word + "\n"
        output_word = output_word + "\t"*tab_count
        tab_count = tab_count + 1
    output_word = output_word + i
    is_upper = i.isupper()
print(output_word)
```
